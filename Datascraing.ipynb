```python
from requests import get
from bs4 import BeautifulSoup
import re
import sys
from datetime import datetime
from datetime import date
from datetime import timedelta
```

```python
def daterange(start_date, end_date):
  for x in range(int ((end_date-start_date).days)):
    yield start_date + timedelta(x)
```

```python
dates = []
```

```python
 
start_date = date(2020, 3, 1)
end_date = date(2020, 5, 1)
```

```python
for idate in daterange(start_date, end_date):
  dates.append(idate.strftime("%Y-%m-%d"))
```

```python
file1 = open("/content/drive/MyDrive/article2020_3.txt",'w',encoding='utf-8')
file2 = open("/content/drive/MyDrive/summary2020_3.txt",'w', encoding='utf-8')
j=0
for date in dates:
  url = 'https://bangla.bdnews24.com/archive/?date='+date
  response = get(url)
  
  html_soup = BeautifulSoup(response.content,'lxml')
  #print(html_soup)
  #summary = html_soup.find_all('h5',{"class": "print-only"})
  
  document_prev = html_soup.find_all(class_='article')
  #print(document_prev)
  print("***********************************************************")
  for i in document_prev:
    #print(i.a.get('href'))
    print("count "+str(j))
    #print(i.get('href'))
    initial_response = get(i.a.get('href'))
 
    intial_html_soup = BeautifulSoup(initial_response.text ,'html.parser' )
    
    processed_text = ""
 
    summary = intial_html_soup.find_all('h5', {"class":"print-only"})
 
    document = intial_html_soup.find_all('div', {"class":"custombody print-only"})
 
    for t in document :
      mytext = t.find_all('p')
      for k in mytext :
        words = k.text.split()
        for wd in words:
          processed_text+=" "+wd
    if len(processed_text)<10:
      continue
    processed_sum=""
    for sum in summary:
      mysum = sum.text.split()
      for xd in mysum:
        processed_sum+=" "+xd
 
    if len(processed_sum)< 4:
      continue
 
    file1.write(processed_text+"\n")
    file2.write(processed_sum+"\n")
    j = j+1
 
 
  print(date)       
    
file1.close()
file2.close()
```
